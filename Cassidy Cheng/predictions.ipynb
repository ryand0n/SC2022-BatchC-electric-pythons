```python
def convert(years):
    if years == 0:
        return 'NO EXPERIENCE'
    elif 1 <= years <= 3:
        return 'LITTLE EXPERIENCE'
    elif 4 <= years <= 6:
        return 'MEDIUM EXPERIENCE'
    else:
        return 'VERY EXPERIENCED'
```

```python
top_15 = list(df['company'].value_counts().head(15).index)
top_15
```

```python
userinput = ['state', 'title', 'race', 'gender', 'experience', 'education', 'in_top_15']
filler2 = df.groupby(by = userinput)['yearsatcompany', 'basesalary', 'stockgrantvalue', 'bonus'].mean()
filler2.reset_index(inplace = True)
filler2.head()
```

```python
def findfillerdata(state, title, race, gender, company, education, yearsofexperience):
    top_15_data = None
    if company in top_15:
        top_15_data = True
    else:
        top_15_data = False
    
    experience = convert(yearsofexperience)
    fillerdata = filler2[(filler2['state'] == state) & 
                     (filler2['title'] == title) & 
                     (filler2['race'] == race) & 
                     (filler2['gender'] == gender) & 
                     (filler2['in_top_15'] == top_15_data) & 
                     (filler2['education'] == education) & 
                     (filler2['experience'] == experience)]
    if len(fillerdata) == 0:
        return 'sorry no data found'
    else:
        return fillerdata
```

```python
findfillerdata('WA', 'Software Engineer', 'Asian', 'Female', 'Google', 'Masters', 2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
```

```python
numeric_features = ['yearsofexperience', 'yearsatcompany', 'basesalary', 'stockgrantvalue', 'bonus', 'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree', 'Highschool', 'Some_College', 'Race_Asian', 'Race_White', 'Race_White', 'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic', 'in_top_15']
numeric_transformer = FunctionTransformer(np.square)
```

```python
categorical_features = ['title', 'gender', 'state']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
preprocessor = ColumnTransformer(transformers = [('num', numeric_transformer, numeric_features), ('cat', categorical_transformer, categorical_features)])
```

```python
clf = Pipeline(
    steps=[('preprocessor', preprocessor),('classifier', KNeighborsRegressor())]
)
```

```python
x = df.loc[:, (df.columns !='totalyearlycompensation')]
y = df.totalyearlycompensation
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
